```python
import pandas as pd
from pathlib import Path


# Load Plankton data
plankton_path = Path("C:/Users/mikek/One Drive New/OneDrive/Leviathan/Ocean/England_1/Plankton_England_1.csv")
df = pd.read_csv(plankton_path)

# Assign points using custom rules: Win = 1, Draw = 0.5, Loss = 0
def custom_points(result):
    if result == 'H':
        return pd.Series([1.0, 0.0])
    elif result == 'A':
        return pd.Series([0.0, 1.0])
    elif result == 'D':
        return pd.Series([0.5, 0.5])
    else:
        return pd.Series([None, None])

df[['HomePoints', 'AwayPoints']] = df['FTR'].apply(custom_points)

# Display key info: season, date, teams, score, result, points
display_columns = ['Season', 'Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HomePoints', 'AwayPoints']
print(df[display_columns].head(6))

# Save updated file
output_path = Path("C:/Users/mikek/One Drive New/OneDrive/Leviathan/Armory/CrowsNest/match_points_custom.csv")
df.to_csv(output_path, index=False)

# Show top 6 rows
df.head(6)
```
